# Visualisation for all graphs

```python
import matplotlib.pyplot as plt
import pandas as pd
import glob
plt.rcParams.update({
  "text.usetex": True,
  "font.family": "Arial"
})
```

```python
def set_dataset(filename):
    data = []
    for file in glob.glob(filename + "/computation*.csv"):
        data.append(pd.read_csv(file, delimiter=',', ))
    data = pd.concat(data)
    data = data.reset_index(drop=True)
    return data
```

```python
data = set_dataset("data")
bdata = set_dataset("bdata")
rdata = set_dataset("rdata")
```
